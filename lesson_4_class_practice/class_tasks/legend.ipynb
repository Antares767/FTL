```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ts = np.linspace(0, 10, 10)

omega = 0.1
xs = np.cos(omega * ts)
ys = np.sin(omega * ts)

plt.plot(xs, ys, 'b-o', label='points')
plt.plot(xs[0], ys[0], 'ro', label='start')
plt.plot(xs[-1], ys[-1], 'go', label='end')
plt.grid()
plt.legend()
plt.show()

```
