```python
import math
import matplotlib.pyplot as plt

class Robot:
    def __init__(self, name):
        self.name = name
        self.x = 0
        self.y = 0
        self.direction = 0   
        self.speed = 0
        self.dt = 0.1
        self.history = [(self.x, self.y)] 
        self.angular_speed = 0 
    
    def get_position(self):
        return (self.x, self.y)
    
    def get_direction(self):
        return self.direction
    
    def get_speed(self):
        return self.speed
    
    def set_speed(self, new_speed):
        if new_speed < 0:
            self.speed = 0
        elif new_speed > 100:
            self.speed = 100
        else:
            self.speed = new_speed

    def set_direction(self, new_direction):
        self.direction = new_direction % 360
    
    def info(self):
        return (f"Робот: {self.name}, "
                f"Положение: {self.get_position()}, "
                f"Направление: {self.get_direction()}, "
                f"Скорость: {self.get_speed()}")

    def move_forward(self, distance):
      
        angle_rad = math.radians(self.direction)
        

        self.x += distance * math.cos(angle_rad)
        self.y += distance * math.sin(angle_rad)
        
      
        self.history.append((self.x, self.y))
    
        pass

    def turn(self, angle):
        self.direction = (self.direction + angle) % 360
    
    def draw_square(self, side_length):
        for _ in range(4):
            self.move_forward(side_length)
            self.turn(90)
    def update(self):

        turn_angle = self.angular_speed * self.dt
        self.turn(turn_angle)  

   
        distance = self.speed * self.dt
        self.move_forward(distance)  # двигаемся вперед

    def simulate(self, duration):

        if duration <= 0:
            print(f"❌ Некорректная длительность симуляции: {duration}")
            return
    
  
        steps = int(duration / self.dt)
    
        for _ in range(steps):
            self.update()
    def draw_path(self):
        
        if not self.history or len(self.history) < 2:
            print(f" {self.name}: Недостаточно точек для отображения пути!")
            return
        
        # Извлекаем координаты
        x_coords = [pos[0] for pos in self.history]
        y_coords = [pos[1] for pos in self.history]
        
     
        plt.figure(figsize=(8, 8))
        plt.plot(x_coords, y_coords, 'b-o')  
        plt.title(f"Путь робота {self.name}")
        plt.xlabel("X координата")
        plt.ylabel("Y координата")
        plt.grid(True)
        plt.axis('equal')
        plt.show()
    def get_angular_speed(self):
       
        return self.angular_speed
        pass
    def set_angular_speed(self, new_angular_speed):
        """Сеттер: устанавливает angular_speed в диапазоне [-180, 180]"""
        # Ограничиваем значение
        self.angular_speed = max(-180, min(180, new_angular_speed))
        pass
```

```python
import math
import matplotlib.pyplot as plt

class Robot:
    def __init__(self, name):
        self.name = name
        self.x = 0
        self.y = 0
        self.direction = 0   
        self.speed = 0
        self.angular_speed = 0
        self.dt = 0.1
        self.history = [(self.x, self.y)]  

    # --- Геттеры ---
    def get_position(self):
        return (self.x, self.y)
    
    def get_direction(self):
        return self.direction
    
    def get_speed(self):
        return self.speed
    
    def get_angular_speed(self):
        return self.angular_speed

    # --- Сеттеры ---
    def set_speed(self, new_speed):
        self.speed = max(0, min(100, new_speed))

    def set_direction(self, new_direction):
        self.direction = new_direction % 360

    def set_angular_speed(self, new_angular_speed):
        self.angular_speed = max(-180, min(180, new_angular_speed))

    # --- Информация ---
    def info(self):
        return (f"Робот: {self.name}, "
                f"Положение: {self.get_position()}, "
                f"Направление: {self.get_direction()}, "
                f"Скорость: {self.get_speed()}")

    # --- Движение ---
    def move_forward(self, distance):
        angle_rad = math.radians(self.direction)
        self.x += distance * math.cos(angle_rad)
        self.y += distance * math.sin(angle_rad)
        self.history.append((self.x, self.y))

    def turn(self, angle):
        self.direction = (self.direction + angle) % 360

    # --- Квадрат ---
    def draw_square(self, side_length):
        for _ in range(4):
            self.move_forward(side_length)
            self.turn(90)

    # --- Обновление ---
    def update(self):
        # Поворот
        turn_angle = self.angular_speed * self.dt
        self.turn(turn_angle)
        # Движение
        distance = self.speed * self.dt
        self.move_forward(distance)

    # --- Симуляция ---
    def simulate(self, duration):
        if duration <= 0:
            print(f"❌ Некорректная длительность симуляции: {duration}")
            return
        steps = int(duration / self.dt)
        for _ in range(steps):
            self.update()

    # --- Визуализация пути ---
    def draw_path(self):
        if not self.history or len(self.history) < 2:
            print(f"{self.name}: Недостаточно точек для отображения пути!")
            return
        x_coords = [pos[0] for pos in self.history]
        y_coords = [pos[1] for pos in self.history]

        plt.figure(figsize=(8, 8))
        plt.plot(x_coords, y_coords, 'b-o', label="Путь робота")
        plt.grid(True)
        plt.axis('equal')
        plt.title(f"Путь робота {self.name}")
        plt.xlabel("X координата")
        plt.ylabel("Y координата")
        plt.show()

    # --- Движение по кругу с визуализацией ---
    def move_in_circle(self, radius, period):
        if radius <= 0 or period <= 0:
            print("❌ Некорректные значения radius или period")
            return

        # Вычисляем параметры движения
        circumference = 2 * math.pi * radius
        linear_speed = circumference / period
        self.set_speed(linear_speed)

        angular_speed = 360 / period
        self.set_angular_speed(angular_speed)

        print(f"⚙️ {self.name} настроен на движение по кругу радиусом {radius} за {period} сек.")
        print(f"   Линейная скорость: {linear_speed:.2f}, Угловая скорость: {angular_speed:.2f}")

        # Симуляция движения по кругу
        self.simulate(period)

        # Визуализация пути с выделением старта и финиша
        if self.history and len(self.history) >= 2:
            x_coords = [pos[0] for pos in self.history]
            y_coords = [pos[1] for pos in self.history]

            plt.figure(figsize=(8, 8))
            plt.plot(x_coords, y_coords, 'b-o', label="Путь робота")

            # Стартовая точка — большая красная
            plt.scatter(x_coords[0], y_coords[0], color='red', s=150, zorder=5)
            plt.text(x_coords[0], y_coords[0]+0.5, "Start", color='red', fontsize=10, ha='center')

            # Финальная точка — зеленая поменьше
            plt.scatter(x_coords[-1], y_coords[-1], color='green', s=100, zorder=5)
            plt.text(x_coords[-1]+0.5, y_coords[-1], "Finish", color='green', fontsize=10, va='center')

            plt.title(f"Путь робота {self.name} по кругу")
            plt.xlabel("X координата")
            plt.ylabel("Y координата")
            plt.grid(True)
            plt.axis('equal')
            plt.legend()
            plt.show()
```

```python
# ПРОВЕРКА: Движение по кругу
robot = Robot("Кружащийся")
robot.move_in_circle(radius=20, period=4.0)

print("--- Настройки для движения по кругу ---")
# Ожидаем: speed = (2 * pi * 20) / 4 = 10 * pi = 31.4
print(f"Расчетная скорость: {robot.get_speed():.2f}")
# Ожидаем: angular_speed = 360 / 4 = 90
print(f"Расчетная угловая скорость: {robot.get_angular_speed():.2f}")
print("-" * 20)


print(f"Стартовая позиция: {robot.get_position()}")

# Симуляция полного круга
robot.simulate(4.0)

pos = robot.get_position()
print(f"Финальная позиция: ({pos[0]:.2f}, {pos[1]:.2f})") # Должен вернуться близко к (0, 0)

# Рисуем траекторию, чтобы убедиться, что это был круг
robot.draw_path()
```

```python
# 1. Создаем робота
final_robot = Robot("Чемпион")

# 2. Задаем параметры
side_length = 20
move_speed = 25           # линейная скорость при движении по стороне
turn_angular_speed = 60   # угловая скорость при повороте на угол
n_sides = 6               # количество сторон шестиугольника

# 3. Рассчитываем времена
turn_angle = 360 / n_sides         # внешний угол шестиугольника
t_move = side_length / move_speed  # время движения по одной стороне
t_turn = turn_angle / turn_angular_speed  # время поворота на угол

# 4. Цикл для рисования шестиугольника
for _ in range(n_sides):
    # Движение по стороне
    final_robot.set_speed(move_speed)
    final_robot.set_angular_speed(0)
    final_robot.simulate(t_move)
    
    # Остановка и поворот
    final_robot.set_speed(0)
    final_robot.set_angular_speed(turn_angular_speed)
    final_robot.simulate(t_turn)

# 5. Сброс скоростей
final_robot.set_speed(0)
final_robot.set_angular_speed(0)

# 6. Визуализация пути робота
if final_robot.history and len(final_robot.history) >= 2:
    x_coords = [pos[0] for pos in final_robot.history]
    y_coords = [pos[1] for pos in final_robot.history]

    plt.figure(figsize=(8, 8))

    # Основной путь
    plt.plot(x_coords, y_coords, 'b-o', label="Путь робота")

    # Стартовая точка — большая красная
    plt.scatter(x_coords[0], y_coords[0], color='red', s=150, zorder=5)
    # Финальная точка — зеленая поменьше
    plt.scatter(x_coords[-1], y_coords[-1], color='green', s=100, zorder=5)

    plt.title(f"Путь робота {final_robot.name}")
    plt.xlabel("X координата")
    plt.ylabel("Y координата")
    plt.grid(True)
    plt.axis('equal')

    # Легенда справа сверху
    plt.scatter([], [], color='red', s=150, label='Старт')   # фиктивная точка для легенды
    plt.scatter([], [], color='green', s=100, label='Финиш')
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

    plt.show()
```
