```python
import math
import matplotlib.pyplot as plt


class Robot:
    
    def __init__(self, name):
        self.name = name
        self.x = 0
        self.y = 0
        self.dt = 0.001
        self.direction = 0
        self.speed = 0
        self.angular_speed = 0
        self.history = [(0, 0)]
    
    def get_position(self):
        return self.x, self.y
    
    def get_direction(self):
        return self.direction
    
    def get_speed(self):
        return self.speed

    def get_angular_speed(self):
        return self.angular_speed
    
    def set_direction(self, new_direction):
        self.direction = new_direction % 360
    
    def set_speed(self, new_speed):
        self.speed = max(0, min(new_speed, 100))

    def set_angular_speed(self, new_angular_speed):
        self.angular_speed = max(-180, min(new_angular_speed, 180))

    def info(self):
        return f"{self.name}: позиция {self.x, self.y} направление {self.direction} скорость {self.speed}"

    def move_forward(self, distance):
        direction_radians = math.radians(self.direction)
        self.x += distance * math.cos(direction_radians)
        self.y += distance * math.sin(direction_radians)
        self.history.append((self.x, self.y))

    def turn(self, angle):
        self.direction += angle
        self.direction %= 360

    def draw_square(self, side_length):
        for i in range(4):
            self.move_forward(side_length)
            self.turn(90)
            print(f"После {i+1}-й стороны: {self.get_position()}")

    def update(self):
        turn_angle = self.angular_speed * self.dt
        distance_per_update = self.speed * self.dt
        self.turn(turn_angle)
        self.move_forward(distance_per_update)

    def simulate(self, duration):
        N = int (duration / self.dt)
        for i in range(N):
            self.update()
    
    def draw_path(self):
        if len(self.history) <= 1:
            print("График невозможно создать")
        else:
            x_coords = [pos[0] for pos in self.history]
            y_coords = [pos[1] for pos in self.history]
            plt.plot(x_coords, y_coords, 'b-', label='points', linewidth=2)
            
            plt.plot(x_coords[0], y_coords[0], 'ro', label='start', markersize=15, markeredgewidth=2)

            plt.plot(x_coords[-1], y_coords[-1], 'go', label='end', markersize=7, markeredgewidth=2)
            
            plt.grid(True)
            plt.legend()
            plt.axis('equal')
    
    def move_in_circle(self, radius, period):
        distance = 2*radius*math.pi
        speed = distance/period
        angular_speed = 360/period
        self.set_speed(speed)
        self.set_angular_speed(angular_speed)
        self.simulate(period)

    def draw_polygon(self, n, side_length):

        self.x = 0
        self.y = 0
        self.direction = 0
        self.history = [(0, 0)]
        
        angle = 360 / n
        for i in range(n):
            self.move_forward(side_length)
            self.turn(angle)


plt.figure(figsize=(10, 10))

# Первый график - треугольник
plt.subplot(2, 2, 1)
robot1 = Robot('robot1')
robot1.draw_polygon(3, 5)
robot1.draw_path()
plt.title("Треугольник (n=3)")

# Второй график - квадрат
plt.subplot(2, 2, 2)
robot2 = Robot('robot2')
robot2.draw_polygon(4, 5)
robot2.draw_path()
plt.title("Квадрат (n=4)")

# Третий график - пятиугольник
plt.subplot(2, 2, 3)
robot3 = Robot('robot3')
robot3.draw_polygon(5, 5)
robot3.draw_path()
plt.title("Пятиугольник (n=5)")

# Четвертый график - восьмиугольник
plt.subplot(2, 2, 4)
robot4 = Robot('robot4')
robot4.draw_polygon(8, 5)
robot4.draw_path()
plt.title("Восьмиугольник (n=8)")

plt.tight_layout() 
plt.show()
```
