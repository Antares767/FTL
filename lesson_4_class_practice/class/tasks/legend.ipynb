```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ts = np.linspace(0, 10, 10)

omega = 2*np.pi
xs = np.cos(omega * ts)
ys = np.sin(omega * ts)

plt.plot(xs, ys, 'b-o', label='points')
plt.plot(xs[0], ys[0], 'ro', label='start', markersize=10)
plt.plot(xs[-1], ys[-1], 'go', label='end')
plt.axis('equal')
plt.grid()
plt.legend()
plt.show()

```

```python
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.plot(xs, ys, 'b-o', label='points')
plt.plot(xs[0], ys[0], 'ro', label='start', markersize=10)
plt.plot(xs[-1], ys[-1], 'go', label='end')
plt.axis('equal')
plt.grid()
plt.legend()

xs_1 = xs / 2
ys_1 = ys
plt.subplot(2, 2, 2)
plt.plot(xs_1, ys_1, 'b-o', label='points')
plt.plot(xs_1[0], ys_1[0], 'ro', label='start', markersize=10)
plt.plot(xs_1[-1], ys_1[-1], 'go', label='end')
plt.axis('equal')
plt.grid()
plt.legend()


xs_1 = xs 
ys_1 = ys / 2
plt.subplot(2, 2, 3)
plt.plot(xs_1, ys_1, 'b-o', label='points')
plt.plot(xs_1[0], ys_1[0], 'ro', label='start', markersize=10)
plt.plot(xs_1[-1], ys_1[-1], 'go', label='end')
plt.axis('equal')
plt.grid()
plt.legend()


xs_1 = xs / 3
ys_1 = ys / 4
plt.subplot(2, 2, 4)
plt.plot(xs_1, ys_1, 'b-o', label='points')
plt.plot(xs_1[0], ys_1[0], 'ro', label='start', markersize=10)
plt.plot(xs_1[-1], ys_1[-1], 'go', label='end')
plt.axis('equal')
plt.grid()
plt.legend()
```
